Read_counts_per_peak_stranded.txt

```python
import pandas as pd

# Laad beide bestanden
df1 = pd.read_csv("/home/ldufour/Documents/Project_Yeast/Output/Quantification_Statistics/read_counts_per_peak_stranded.txt", sep="\t", comment="#")
df2 = pd.read_csv("/home/saif/Documents/Saif_project_Yeast/approach2/outputs/read_counts_per_peak_stranded.txt", sep="\t", comment="#")

# Kolomnamen
print("=== Column ===")
print(f"File Lara:   {df1.columns.tolist()}")
print(f"File Saif: {df2.columns.tolist()}")

# number of rows
print("\n=== Number of rows  ===")
print(f"File Lara:   {df1.shape}")
print(f"File Saif: {df2.shape}")

# Rename
count_col1 = df1.columns[-1]
count_col2 = df2.columns[-1]
df1 = df1.rename(columns={count_col1: "count"})
df2 = df2.rename(columns={count_col2: "count"})

# difference counts on GeneID
merged = pd.merge(df1[["Geneid", "count"]], 
                  df2[["Geneid", "count"]], 
                  on="Geneid", suffixes=("_ldufour", "_saif"))

merged["difference"] = merged["count_ldufour"] - merged["count_saif"]

print("\n=== statistical differences ===")
print(merged["difference"].describe())

print(f"\nNumber of rows different: {(merged['difference'] != 0).sum()}")
print(f"Number of row no difference: {(merged['difference'] == 0).sum()}")

print("\n=== first 20 difference ===")
print(merged[merged["difference"] != 0].head(20))
```

```python
import pandas as pd

# === File Paths ===
ldufour_path = "/home/ldufour/Documents/Project_Yeast/Output/Quantification_Statistics/filtered_stranded_peaks_expanded_and_tagged.csv"
saif_path    = "/home/saif/Documents/Saif_project_Yeast/approach2/outputs/filtered_stranded_peaks_expanded_and_tagged.csv"  

# === Load Data ===
df_l = pd.read_csv(ldufour_path)
df_s = pd.read_csv(saif_path)

# === Info ===
print("=== COLUMNS ===")
print(f"Lara's file:  {df_l.columns.tolist()}")
print(f"Saif's file:  {df_s.columns.tolist()}")

print("\n=== SHAPE ===")
print(f"Lara's file:  {df_l.shape}")
print(f"Saif's file:  {df_s.shape}")

# === Merge on GeneID (unique per row due to tagging) ===
merged = df_l.merge(df_s, on='GeneID', suffixes=('_lara', '_saif'))

# === Compare numerical columns ===
for col in ['negative', 'positive', 'Total', 'Positive_pct', 'Negative_pct']:
    merged[f'diff_{col}'] = merged[f'{col}_lara'] - merged[f'{col}_saif']

# === Compare categorical columns ===
merged['diff_BasePeakID']      = merged['BasePeakID_lara']      != merged['BasePeakID_saif']
merged['diff_Assigned_Strand'] = merged['Assigned_Strand_lara'] != merged['Assigned_Strand_saif']

# === Statistics ===
diff_cols = [f'diff_{col}' for col in ['negative', 'positive', 'Total', 'Positive_pct', 'Negative_pct']]

print("\n=== DIFFERENCE STATISTICS (numerical) ===")
print(merged[diff_cols].describe())

print("\n=== CATEGORICAL DIFFERENCES ===")
print(f"Number of rows with different BasePeakID:      {merged['diff_BasePeakID'].sum()}")
print(f"Number of rows with different Assigned_Strand: {merged['diff_Assigned_Strand'].sum()}")

print("\n=== ROWS WITH DIFFERENCE ===")
has_diff = merged[
    (merged[diff_cols] != 0).any(axis=1) |
    merged['diff_BasePeakID'] |
    merged['diff_Assigned_Strand']
]
print(f"Number of rows with difference:    {len(has_diff)}")
print(f"Number of rows without difference: {len(merged) - len(has_diff)}")

print("\n=== FIRST 20 ROWS WITH DIFFERENCE ===")
print(has_diff[['GeneID'] + diff_cols + ['diff_BasePeakID', 'diff_Assigned_Strand']].head(20).to_string())
```

summary1_with_correct_read_counts.csv

```python
import pandas as pd

# === File Paths ===
ldufour_path = "/home/ldufour/Documents/Project_Yeast/Output/Quantification_Statistics/summary1_with_correct_read_counts.csv"
saif_path    = "/home/saif/Documents/Saif_project_Yeast/approach2/outputs_approach2/Quantification_Statistics/summary1_with_correct_read_counts.csv"  

# === Load Data ===
df_l = pd.read_csv(ldufour_path)
df_s = pd.read_csv(saif_path)

# === Info ===
print("=== COLUMNS ===")
print(f"Lara's file:  {df_l.columns.tolist()}")
print(f"Saif's file:  {df_s.columns.tolist()}")

print("\n=== SHAPE ===")
print(f"Lara's file:  {df_l.shape}")
print(f"Saif's file:  {df_s.shape}")

# === Merge on transcript_id (unique per row) ===
merged = df_l.merge(df_s, on='transcript_id', suffixes=('_lara', '_saif'))

# === Compare numerical columns ===
for col in ['overlap_nt', 'cds_length', 'overlap_pct', 'transcript_length', 'read_counts']:
    merged[f'diff_{col}'] = merged[f'{col}_lara'] - merged[f'{col}_saif']

# === Compare categorical columns ===
for col in ['gene_id', 'is_longest', 'transcript_strand', 'BasePeakID']:
    merged[f'diff_{col}'] = merged[f'{col}_lara'] != merged[f'{col}_saif']

# === Statistics ===
diff_num_cols = [f'diff_{col}' for col in ['overlap_nt', 'cds_length', 'overlap_pct', 'transcript_length', 'read_counts']]
diff_cat_cols = [f'diff_{col}' for col in ['gene_id', 'is_longest', 'transcript_strand', 'BasePeakID']]

print("\n=== DIFFERENCE STATISTICS (numerical) ===")
print(merged[diff_num_cols].describe())

print("\n=== CATEGORICAL DIFFERENCES ===")
for col in diff_cat_cols:
    print(f"Number of rows with different {col}: {merged[col].sum()}")

print("\n=== ROWS WITH DIFFERENCE ===")
has_diff = merged[
    (merged[diff_num_cols] != 0).any(axis=1) |
    merged[diff_cat_cols].any(axis=1)
]
print(f"Number of rows with difference:    {len(has_diff)}")
print(f"Number of rows without difference: {len(merged) - len(has_diff)}")

print("\n=== FIRST 20 ROWS WITH DIFFERENCE ===")
print(has_diff[['transcript_id'] + diff_num_cols + diff_cat_cols].head(20).to_string())
```

```python
# === Inspect read_counts differences ===
read_diff = merged[merged['diff_read_counts'] != 0][['transcript_id', 'read_counts_lara', 'read_counts_saif', 'diff_read_counts']]

print("=== READ COUNTS DIFFERENCE SUMMARY ===")
print(f"Rows with difference:    {len(read_diff)}")
print(f"Rows where Lara > Saif:  {(read_diff['diff_read_counts'] > 0).sum()}")
print(f"Rows where Saif > Lara:  {(read_diff['diff_read_counts'] < 0).sum()}")

print("\n=== DISTRIBUTION OF DIFFERENCES ===")
print(read_diff['diff_read_counts'].describe())

print("\n=== LARGEST DIFFERENCES ===")
print(read_diff.reindex(read_diff['diff_read_counts'].abs().sort_values(ascending=False).index).head(20).to_string())

print("\n=== ROWS WHERE LARA = 0 BUT SAIF != 0 ===")
zero_lara = merged[(merged['read_counts_lara'] == 0) & (merged['read_counts_saif'] != 0)]
print(f"Count: {len(zero_lara)}")
print(zero_lara[['transcript_id', 'read_counts_lara', 'read_counts_saif']].head(10).to_string())

print("\n=== ROWS WHERE SAIF = 0 BUT LARA != 0 ===")
zero_saif = merged[(merged['read_counts_saif'] == 0) & (merged['read_counts_lara'] != 0)]
print(f"Count: {len(zero_saif)}")
print(zero_saif[['transcript_id', 'read_counts_lara', 'read_counts_saif']].head(10).to_string())
```
